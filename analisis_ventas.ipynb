# 📊 Análisis de Ventas 2024
Análisis exploratorio básico usando Python y Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear datos de ventas
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'],
    'Ventas': [1500, 1800, 1200, 2100, 1900, 2400],
    'Gastos': [800, 950, 700, 1100, 1000, 1200]
}

df = pd.DataFrame(datos)
print(df)
```

```python
# Estadísticas básicas
print("📈 Resumen estadístico:")
print(df.describe())
print(f"\n✅ Mes con más ventas: {df.loc[df['Ventas'].idxmax(), 'Mes']}")
print(f"❌ Mes con menos ventas: {df.loc[df['Ventas'].idxmin(), 'Mes']}")
```

```python
# Gráfico de ventas vs gastos
plt.figure(figsize=(10, 5))
plt.plot(df['Mes'], df['Ventas'], marker='o', color='blue', label='Ventas')
plt.plot(df['Mes'], df['Gastos'], marker='o', color='red', label='Gastos')
plt.title('Ventas vs Gastos 2024')
plt.xlabel('Mes')
plt.ylabel('Monto ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 📝 Conclusiones
- El mes con mayor venta fue **Junio** con $2,400
- El mes con menor venta fue **Marzo** con $1,200
- La tendencia general de ventas es **positiva**
- Los gastos se mantienen estables respecto a las ventas
